````python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import nest
import nest.topology as topp
````

# layers

````python
""" on-grid"""
ayer_dict_ex = {"extent" : [2.,2.], # size (mm)
                 "rows" : 10, 
                 "columns" : 10, 
                 "elements" : "iaf_psc_alpha"} # the element at each (x,y) coordinate in the grid (homo)
layer_on_grid = topp.CreateLayer(ayer_dict_ex)
````

````python
""" off grid"""
# grid with jitter
jit = 0.03
xs = np.arange(-0.5,.501,0.1)
poss = np.array(list(itertools.product(xs,xs)))
poss += np.random.uniform(-jit,jit,poss.shape)
layer_dict_ex = {"positions": list(poss),
        "extent" : [1.1,1.1],
        "elements" : "iaf_psc_alpha"}
layer_off_grid = topp.CreateLayer(layer_dict_ex)
````

````python
""" composite layers (hetero.)"""
nest.CopyModel("iaf_psc_alpha","pyramids") # pyramidal neuron
nest.CopyModel("iaf_psc_alpha","intern", {"V_th": -52.}) # inhi. interneuron
comp_layer = topp.CreateLayer({"rows":5,"columns":5, 
        "elements": ["pyramids",4, "intern", "poisson_generator", "noise_generator"]}) 
# network consisting 4 pyramidal cells and an interneuron, a poisson gen and a noise gen
````

# Connection dictionary
```connection_type``` and ```mask``` are mandatory parameters.

```connection_type```: how nodes are selected (divergent,convergent)
```mask```: which neurons are going to be selected as an target (which shape)

````python
conn_dict_ex = conn_dict_ex = {"connection_type":"divergent","mask":{"circular":{"radius":0.5}}}
````

## Connect

````python
topp.ConnectLayers(layer_off_grid, layer_on_grid, conn_dict_ex)
````
